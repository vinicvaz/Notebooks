```python
from shapely.geometry import Polygon
import click
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from descartes.patch import PolygonPatch

%matplotlib inline
```

## Figura  a ser calculada
  
**Figura 3x3**

```python
def inferior(x):
    return (x**3)/9
    
```

```python
x_sup = np.linspace(0.0, 3.0, 200)
x_inf = np.flipud(x_sup)
x = y = np.linspace(0.0, 3.0, 20) ## vetor de coordenadas para ajustar o plot da figura
```

```python
coord_sup= [(xx, yy) for xx, yy in zip(x_sup, x_sup)]
coord_inf= [(xx, yy) for xx, yy in zip(x_inf, inferior(x_inf))]

coords = [*coord_sup, *coord_inf]
```

```python
polygon = Polygon(coords)
```

```python
aa = polygon.area
cc = polygon.centroid.coords.xy
xe, ye = polygon.exterior.coords.xy
print("Cordenadas Centroide:")
print("X centroide: ", round(cc[0][0],4))
print("Y centroide: ", round(cc[1][0],4))
```

```python
fig, ax = plt.subplots(figsize=(12,8), dpi=60)
patch = PolygonPatch(polygon, fc='black',alpha=0.3, zorder=0)
ax.add_patch(patch)
ax.plot(xe, ye, color='black', linewidth=1.5, linestyle='-', marker='', zorder=0)
ax.axhline(0, color='black', linestyle='--')
ax.axvline(0, color='black', linestyle='--')
ax.scatter(cc[0], cc[1], color='red', s=50, marker='o', zorder=5)
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)
plt.xticks(x)
plt.yticks(y)
```
